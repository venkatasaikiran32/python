4.Create a simple banking system using object-oriented concepts in Python. Design classes representing 

different types of bank accounts such as savings and checking. Implement methods for deposit, withdraw, 

and balance inquiry. Utilize inheritance to manage different account types efficientl

```python
class BankAccount:
    def __init__(self, account_number, account_holder):
        self.account_number = account_number
        self.account_holder = account_holder
        self.balance = 0.0

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited ${amount:.3f} to account {self.account_number}.")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew ${amount:.3f} from account {self.account_number}.")
        else:
            print("Withdrawal amount must be positive and less than or equal to the current balance.")

    def get_balance(self):
        return self.balance

    def __str__(self):
        return f"Account Number: {self.account_number}, Account Holder: {self.account_holder}, Balance: ${self.balance:.2f}"


class SavingsAccount(BankAccount):
    def __init__(self, account_number, account_holder, interest_rate):
        super().__init__(account_number, account_holder)
        self.interest_rate = interest_rate

    def add_interest(self):
        interest = self.balance * self.interest_rate / 100
        self.balance += interest
        print(f"Interest of ${interest:.2f} added to account {self.account_number}.")

    def __str__(self):
        return f"Savings {super().__str__()}, Interest Rate: {self.interest_rate}%"


class CheckingAccount(BankAccount):
    def __init__(self, account_number, account_holder, overdraft_limit):
        super().__init__(account_number, account_holder)
        self.overdraft_limit = overdraft_limit

    def withdraw(self, amount):
        if 0 < amount <= self.balance + self.overdraft_limit:
            self.balance -= amount
            print(f"Withdrew ${amount:.2f} from account {self.account_number}.")
        else:
            print("Withdrawal amount must be positive and less than or equal to the current balance plus overdraft limit.")

    def __str__(self):
        return f"Checking {super().__str__()}, Overdraft Limit: ${self.overdraft_limit:.2f}"



    

#testing the sample banking system whether working fine or not

#two accounts
#saving accounts-->which is inheriting the base class -->bank account
#    -->so that they have access to get methods of bank account
#checking account-->it is also inheriting the base class -->bank account 
#--->so that they have access to get methods of bank account
#in checking account we have facility overdraft -->limit on ammount of money allowed 
#to be withdrawn in excess of the credit balance of a bank


savings = SavingsAccount(23456785653, "saikiran ", 1.7)  #account number,account holder,interest rate
checking = CheckingAccount(357252862, "venkat", 500.0)  #account no ,account holder ,overdraft limit

# Depositing  money
savings.deposit(10000)
checking.deposit(5000)

# Withdrawing  money
savings.withdraw(200)
checking.withdraw(600)

#Adding interest to savings account
savings.add_interest()

# getting account details
print(savings)
print(checking)

# Checking balances of the accounts 
print(f"Savings Account Balance: ${savings.get_balance():.2f}")
print(f"Checking Account Balance: ${checking.get_balance():.2f}")
```

```python
class Room:
    _id_counter = 1

    def __init__(self, room_number, room_type, rate):
        self.__id = Room._id_counter
        Room._id_counter += 1
        self.room_number = room_number
        self.room_type = room_type
        self.rate = rate
        self.__availability = True

    def get_id(self):
        return self.__id

    def is_available(self):
        return self.__availability

    def book_room(self):
        if self.__availability:
            self.__availability = False
            print(f"Room {self.room_number} booked successfully.")
        else:
            print(f"Room {self.room_number} is already booked.")

    def check_in(self):
        if not self.__availability:
            print(f"Guest checked into room {self.room_number}.")
        else:
            print(f"Room {self.room_number} is not booked yet.")

    def check_out(self):
        if not self.__availability:
            self.__availability = True
            print(f"Guest checked out of room {self.room_number}.")
        else:
            print(f"Room {self.room_number} is already available.")

    def __str__(self):
        return (f"ID: {self.get_id()}, Room Number: {self.room_number}, Room Type: {self.room_type}, "
                f"Rate: ${self.rate:.2f}, Available: {'Yes' if self.__availability else 'No'}")


class SuiteRoom(Room):
    def __init__(self, room_number, rate, has_jacuzzi):
        super().__init__(room_number, "Suite", rate)
        self.has_jacuzzi = has_jacuzzi

    def __str__(self):
        return (super().__str__() + f", Jacuzzi: {'Yes' if self.has_jacuzzi else 'No'}")


class StandardRoom(Room):
    def __init__(self, room_number, rate, has_view):
        super().__init__(room_number, "Standard", rate)
        self.has_view = has_view

    def __str__(self):
        return (super().__str__() + f", View: {'Yes' if self.has_view else 'No'}")


class Hotel:
    def __init__(self):
        self.rooms = []

    def add_room(self, room):
        self.rooms.append(room)
        print(f"Added room {room.room_number} to the hotel.")

    def find_room_by_id(self, room_id):
        for room in self.rooms:
            if room.get_id() == room_id:
                return room
        return None

    def book_room(self, room_id):
        room = self.find_room_by_id(room_id)
        if room:
            room.book_room()
        else:
            print(f"Room with ID {room_id} not found.")

    def check_in(self, room_id):
        room = self.find_room_by_id(room_id)
        if room:
            room.check_in()
        else:
            print(f"Room with ID {room_id} not found.")

    def check_out(self, room_id):
        room = self.find_room_by_id(room_id)
        if room:
            room.check_out()
        else:
            print(f"Room with ID {room_id} not found.")

    def display_rooms(self):
        for room in self.rooms:
            print(room)


# Example usage
hotel = Hotel()

# Adding rooms to the hotel
suite1 = SuiteRoom(101, 250.00, True)
standard1 = StandardRoom(102, 100.00, False)

hotel.add_room(suite1)
hotel.add_room(standard1)

# Displaying all rooms
print("\nRooms in the hotel:")
hotel.display_rooms()

# Booking a room
hotel.book_room(suite1.get_id())

# Checking in a guest
hotel.check_in(suite1.get_id())

# Displaying all rooms after check-in
print("\nRooms in the hotel after check-in:")
hotel.display_rooms()

# Checking out a guest
hotel.check_out(suite1.get_id())

# Displaying all rooms after check-out
print("\nRooms in the hotel after check-out:")
hotel.display_rooms()
```

```python
class Member:
    _id_counter = 1

    def __init__(self, name, age, membership_type):
        self.__id = Member._id_counter
        Member._id_counter += 1
        self.name = name
        self.age = age
        self.membership_type = membership_type
        self.__membership_status = "Active"

    def get_id(self):
        return self.__id

    def get_membership_status(self):
        return self.__membership_status

    def renew_membership(self):
        self.__membership_status = "Active"
        print(f"Membership for {self.name} has been renewed.")

    def cancel_membership(self):
        self.__membership_status = "Cancelled"
        print(f"Membership for {self.name} has been cancelled.")

    def __str__(self):
        return (f"ID: {self.get_id()}, Name: {self.name}, Age: {self.age}, Membership Type: {self.membership_type}, "
                f"Status: {self.get_membership_status()}")


class FamilyMember(Member):
    def __init__(self, name, age, membership_type, family_size):
        super().__init__(name, age, membership_type)
        self.family_size = family_size

    def __str__(self):
        return super().__str__() + f", Family Size: {self.family_size}"


class IndividualMember(Member):
    def __init__(self, name, age, membership_type):
        super().__init__(name, age, membership_type)

    def __str__(self):
        return super().__str__()


class FitnessClub:
    def __init__(self):
        self.members = []

    def register_member(self, member):
        self.members.append(member)
        print(f"Registered new member: {member.name}")

    def find_member_by_id(self, member_id):
        for member in self.members:
            if member.get_id() == member_id:
                return member
        return None

    def renew_membership(self, member_id):
        member = self.find_member_by_id(member_id)
        if member:
            member.renew_membership()
        else:
            print(f"Member with ID {member_id} not found.")

    def cancel_membership(self, member_id):
        member = self.find_member_by_id(member_id)
        if member:
            member.cancel_membership()
        else:
            print(f"Member with ID {member_id} not found.")

    def display_members(self):
        for member in self.members:
            print(member)


# trsting with one example
club = FitnessClub()

# Registering members
member1 = IndividualMember("sai kiran", 30, "individual")
member2 = FamilyMember("mandapati venkat", 45, "family", 4)

club.register_member(member1)
club.register_member(member2)

# Displaying all members
print("\nMembers in the club:")
club.display_members()

# Renewing a membership
club.renew_membership(member1.get_id())

# Cancelling a membership
club.cancel_membership(member2.get_id())

# Displaying all members after renewal and cancellation
print("\nMembers in the club after renewal and cancellation:")
club.display_members()
```

--> Inherit from the Event class to create a PrivateEvent class and a PublicEvent class, each with their own

```python
import uuid

class Event:
    def __init__(self, name, date, time, location):
        self.__id = uuid.uuid4()  # UUID generates unique id everytime 
        self.name = name
        self.date = date
        self.time = time
        self.location = location
        self.__attendees = []

    def get_id(self):
        return self.__id

    def add_attendee(self, attendee):
        if attendee not in self.__attendees:
            self.__attendees.append(attendee)
            print(f"Added {attendee} to event {self.name}.")
        else:
            print(f"{attendee} is already attending the event {self.name}.")

    def remove_attendee(self, attendee):
        if attendee in self.__attendees:
            self.__attendees.remove(attendee)
            print(f"Removed {attendee} from event {self.name}.")
        else:
            print(f"{attendee} is not attending the event {self.name}.")

    def get_total_attendees(self):
        return len(self.__attendees)

    def __str__(self):
        return (f"ID: {self.get_id()}, Name: {self.name}, Date: {self.date}, Time: {self.time}, "
                f"Location: {self.location}, Attendees: {self.get_total_attendees()}")


class PrivateEvent(Event):
    def __init__(self, name, date, time, location, host_name):
        super().__init__(name, date, time, location)
        self.host_name = host_name

    def __str__(self):
        return super().__str__() + f", Host: {self.host_name}"


class PublicEvent(Event):
    def __init__(self, name, date, time, location, ticket_price):
        super().__init__(name, date, time, location)
        self.ticket_price = ticket_price

    def __str__(self):
        return super().__str__() + f", Ticket Price: ${self.ticket_price:.2f}"


class EventManager:
    def __init__(self):
        self.events = []

    def create_event(self, event):
        self.events.append(event)
        print(f"Created new event: {event.name}")

    def find_event_by_id(self, event_id):
        for event in self.events:
            if event.get_id() == event_id:
                return event
        return None

    def add_attendee(self, event_id, attendee):
        event = self.find_event_by_id(event_id)
        if event:
            event.add_attendee(attendee)
        else:
            print(f"Event with ID {event_id} not found.")

    def remove_attendee(self, event_id, attendee):
        event = self.find_event_by_id(event_id)
        if event:
            event.remove_attendee(attendee)
        else:
            print(f"Event with ID {event_id} not found.")

    def display_events(self):
        for event in self.events:
            print(event)


# test case 
manager = EventManager()

# Creating events
private_event = PrivateEvent("Company Retreat", "2024-10-15", "9:00 AM", "gandhi park", "sai kiran")
public_event = PublicEvent("Summer Festival", "2024-08-10", "2:00 PM", "City Park", 20.00)

manager.create_event(private_event)
manager.create_event(public_event)

# Adding and removing attendees
manager.add_attendee(private_event.get_id(), "vishwa mohan")
manager.add_attendee(private_event.get_id(), "sudhansh kumar")
manager.add_attendee(public_event.get_id(), "sai kumar")
manager.add_attendee(public_event.get_id(), "krish naik")


# Displaying all events
print("\nAll Events:")
manager.display_events()

# Removing an attendee
manager.remove_attendee(private_event.get_id(), "Alice Smith")

# Displaying all events after updates
print("\nAll Events After Updates:")
manager.display_events()
```

10.Write a code for Airline Reservation System using OOps

--> Create a Flight 'lass that has attributes such as flight number, departure and arrival airports, departure and 
arrival times, and available seats (private)

--> Implement methods to book a seat, cancel a reservation, and get the remaining available seats\

--> Use encapsulation to hide the flight's unique identification number

--> Inherit from the Flight class to create a DomesticFlight class and an InternationalFlight class, each with their 
own specific attributes and methods.

```python
import uuid
from datetime import datetime

class Flight:
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats):
        self.__id = uuid.uuid4()  # Generate a unique ID using UUID
        self.flight_number = flight_number
        self.departure_airport = departure_airport
        self.arrival_airport = arrival_airport
        self.departure_time = departure_time
        self.arrival_time = arrival_time
        self.__total_seats = total_seats
        self.__available_seats = total_seats

    def get_id(self):
        return self.__id

    def book_seat(self):
        if self.__available_seats > 0:
            self.__available_seats -= 1
            print(f"Seat booked on flight {self.flight_number}.")
        else:
            print(f"No available seats on flight {self.flight_number}.")

    def cancel_reservation(self):
        if self.__available_seats < self.__total_seats:
            self.__available_seats += 1
            print(f"Reservation cancelled on flight {self.flight_number}.")
        else:
            print(f"No reservations to cancel on flight {self.flight_number}.")

    def get_remaining_seats(self):
        return self.__available_seats

    def __str__(self):
        return (f"ID: {self.get_id()}, Flight Number: {self.flight_number}, Departure: {self.departure_airport} "
                f"at {self.departure_time}, Arrival: {self.arrival_airport} at {self.arrival_time}, "
                f"Available Seats: {self.get_remaining_seats()}")


    
    
    
    
    #it is inheriting Flight base class
    
class DomesticFlight(Flight):
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats, state):
        super().__init__(flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats)
        self.state = state

    def __str__(self):
        return super().__str__() + f", State: {self.state}"

    
    #it is inheriting flight base class
    

class InternationalFlight(Flight):
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats, country):
        super().__init__(flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats)
        self.country = country

    def __str__(self):
        return super().__str__() + f", Destination Country: {self.country}"


class AirlineReservationSystem:
    def __init__(self):
        self.flights = []

    def add_flight(self, flight):
        self.flights.append(flight)
        print(f"Added flight: {flight.flight_number}")

    def find_flight_by_id(self, flight_id):
        for flight in self.flights:
            if flight.get_id() == flight_id:
                return flight
        return None

    def book_seat(self, flight_id):
        flight = self.find_flight_by_id(flight_id)
        if flight:
            flight.book_seat()
        else:
            print(f"Flight with ID {flight_id} not found.")

    def cancel_reservation(self, flight_id):
        flight = self.find_flight_by_id(flight_id)
        if flight:
            flight.cancel_reservation()
        else:
            print(f"Flight with ID {flight_id} not found.")

    def display_flights(self):
        for flight in self.flights:
            print(flight)


# Example usage
system = AirlineReservationSystem()

# Creating flights
domestic_flight = DomesticFlight(
    flight_number="AI202", departure_airport="DEL", arrival_airport="BLR",
    departure_time=datetime(2024, 8, 15, 8, 00), arrival_time=datetime(2024, 8, 15, 10, 00),
    total_seats=180, state="Karnataka"
)

international_flight = InternationalFlight(
    flight_number="AI303", departure_airport="DEL", arrival_airport="LHR",
    departure_time=datetime(2024, 8, 20, 22, 00), arrival_time=datetime(2024, 8, 21, 5, 30),
    total_seats=250, country="United Kingdom"
)

system.add_flight(domestic_flight)
system.add_flight(international_flight)

# Booking and cancelling seats
system.book_seat(domestic_flight.get_id())
system.book_seat(international_flight.get_id())
system.cancel_reservation(domestic_flight.get_id())

# Displaying all flights
print("\nAll Flights:")
system.display_flights()
```
